```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Imersão Python - Tabela de ações - Gráficos.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/Imersão Python - Tabela de ações - Gráficos.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel("/content/Imersão Python - Tabela de ações - Gráficos.xlsx", sheet_name="Ticker")
df_ticker.head(10)
```

```python
df_gemini = pd.read_excel("/content/Imersão Python - Tabela de ações - Gráficos.xlsx", sheet_name="Gemini")
df_gemini.head(10)
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Último (R$)':'Valor_Final', 'Var. Dia (%)':'Var_Dias_Pct'}).copy()
df_principal
```

```python
df_principal['Var_Pct'] = df_principal['Var_Dias_Pct'] / 100
df_principal['Var_Inicial'] = df_principal['Valor_Final'] / (df_principal['Var_Pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on = 'Ativo', right_on = 'Código', how = 'left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao'] = (df_principal['Valor_Final'] - df_principal['Var_Inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Qtde. Teórica':'Qtd_Teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['variacao'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Manteve'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on = 'Ativo', right_on = 'Ticker', how = 'left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_gemini, left_on = 'Nome', right_on = 'Empresa', how = 'left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
# Calculando o maior valor da coluna 'Variacao'
maior = df_principal['variacao'].max()

# Calculando o menor valor da coluna 'Variacao'
menor = df_principal['variacao'].min()

# Calculando a média da coluna 'Variacao'
media = df_principal['variacao'].mean()

# Calculando a média da coluna 'Variacao' apenas para os registros em que 'Resultado' é 'Subiu'
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['variacao'].mean()

# Calculando a média da coluna 'Variacao' apenas para os registros em que 'Resultado' é 'Desceu'
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['variacao'].mean()

# Agora, vamos imprimir os resultados
print(f"Maior: R$ {maior:,.2f}")
print(f"Menor: R$ {menor:,.2f}")
print(f"Média: R$ {media:,.2f}")
print(f"Média de quem Subiu: R$ {media_subiu:,.2f}")
print(f"Média de quem Desceu: R$ {media_desceu:,.2f}")
```
